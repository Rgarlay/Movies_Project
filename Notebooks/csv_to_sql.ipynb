```python
from sqlalchemy import Table, Column, Integer, Float, Boolean, String
```

```python
import pandas as pd

subs = pd.read_csv('subs.csv')

subs.columns
```

```python
subs.sample()
```

```python
content_consumption = pd.read_csv('content_consumption.csv')
content_consumption.sample()
```

```python
meta = MetaData()

liocinema_table = Table("liocinema", meta,
                    Column('user_id', String(50),primary_key=True),
                    Column('age_group', String(20)),
                    Column("city_tier", String(20)),
                    Column("old_subscription_plan",String(20)),
                    Column("new_subscription_plan", String(20)),
                    Column("did_change_plan", Boolean),
                    Column("upgrade_or_not", Boolean)
                    )

content_consumption = Table("cons", meta,
                            Column('user_id', String(50), primary_key=True),
                            Column("device_type", String(20)),
                            Column("total_watch_time_mins", Integer)
                            )

meta.create_all(engine)
```
