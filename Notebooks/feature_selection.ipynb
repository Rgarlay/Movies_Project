```python
import pandas as pd
import numpy as np
```

```python
subs = pd.read_csv("subs.csv")
```

```python
changed_plan = subs[subs['plan_change_date'].notna()]
changed_plan.loc[:,'did_change_plan'] = 1
changed_plan['upgrade_or_not'] = changed_plan.apply(lambda x: 1 
                                                    if x['subscription_plan'] == 'Free'
                                                    or (x['subscription_plan'] == 'Basic' and x['new_subscription_plan'] == 'Premium')
                                                                         else 0,axis = 1)
```

```python
subs_modified = changed_plan.copy()
```
