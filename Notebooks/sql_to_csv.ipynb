Code to export the 3 databases from sql format into csv files.

```python
import pandas as pd
from sqlalchemy import create_engine,inspect
import pymysql
```

```python
engine = create_engine('mysql+pymysql://root:workbench2025@localhost:3306/liocinema_db')
conn = engine.connect()
```

```python
insp = inspect(engine)
print(insp.get_table_names())
```

```python
subs = pd.read_sql_query("select * from subscribers", con=conn)
content_consumption = pd.read_sql_query("select * from content_consumption", con = conn)
contents = pd.read_sql_query("select * from contents", con = conn)
```

```python
subs.sample()
```

```python
subs.to_csv("subs.csv", index=False, header=True)
content_consumption.to_csv("content_consumption.csv", index=False, header=True)
contents.to_csv("contents.csv", header=True, index=False)
```
